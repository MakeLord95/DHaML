# 7 - Linear Regression

```python
# Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
```

```python
# Loading the datafile
df = pd.read_csv("../datasets/tvmarketing.csv")
df.tail()
```

```python
# Split the data
X = df.drop(columns=['Sales'])
y = df['Sales']
```

```python
# Display X head
X.head()
```

```python
# Display y head
y.head()
```

```python
# Display data in a scatter diagram
plt.scatter(X, y)

plt.xlabel('TV Marketing')
plt.ylabel('Sales')

plt.grid(True)
plt.show()
```

```python
# Creating a LinearRegression model
model = LinearRegression().fit(X, y)
```

```python
# Define beta0 and beta1
beta0 = model.intercept_
beta1 = model.coef_
```

```python
# Display beta0
print(f"Beta 0: {beta0:.2f}")
```

```python
# Display beta1
print(f"Beta 1: {beta1[0]:.2f}")
```

```python
# Display beta0 & beta1 in a diagram
xs = np.linspace(X.min(), X.max())
ys = beta0 + beta1 * xs

plt.plot(xs, ys)

plt.scatter(X, y)

plt.xlabel("TV Advertising (1000$)")
plt.ylabel("Sales revenue (M$)")

plt.grid(True)
plt.show()
```

```python
# Create predictions and calculate the mean absolute error
preds = model.predict(X)
mae = mean_absolute_error(y, preds)
print(f"Mean absolute error: {mae:.2f}")
```

```python
# Create a residual plot
plt.scatter(X, preds - y)
plt.grid(True)
plt.show()
```

```python
# Display a histogram
plt.hist(preds - y, bins=10)
plt.grid(True)
plt.show()
```
